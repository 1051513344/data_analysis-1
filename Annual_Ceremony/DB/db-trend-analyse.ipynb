```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Scatter, Line, Funnel
from pyecharts.commons.utils import JsCode
```

```python
# 2019-12 前十名
```

```python
def bar_db() -> Bar:
    c = (
        Bar()
        .add_xaxis(df_201912['name'].values.tolist()[:20])
        .add_yaxis("数据库流行度", df_201912['value'].values.tolist()[:20])
        .set_series_opts(itemstyle_opts={
            "normal": {
                "color": JsCode("""new echarts.graphic.LinearGradient(0, 0, 0, 1, [{
                    offset: 0,
                    color: 'rgba(0, 244, 255, 1)'
                }, {
                    offset: 1,
                    color: 'rgba(0, 77, 167, 1)'
                }], false)"""),
                "barBorderRadius": [30, 30, 30, 30],
                "shadowColor": 'rgb(0, 160, 221)',
            }},
                        label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title=""),
                        #datazoom_opts=opts.DataZoomOpts(),
                        xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-30)),)
    )
    return c
```

```python
bar_db().render_notebook()
```

```python
def scatter_db() -> Scatter:
    c = (
        Scatter()
        .add_xaxis(df_201912['name'].values.tolist()[:20])
        .add_yaxis("数据库流行度", df_201912['value'].values.tolist()[:20])
        .set_global_opts(
            title_opts=opts.TitleOpts(title=""),
            visualmap_opts=opts.VisualMapOpts(type_="size", max_=1500, min_=50),
            xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-30)),
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    )
    return c
```

```python
scatter_db().render_notebook()
```

```python
oracle_data = df_new[df_new['name'] == 'Oracle']
mysql_data = df_new[df_new['name'] == 'MySQL']
MSSQL_data = df_new[df_new['name'] == 'Microsoft SQL Server']
Pg_data = df_new[df_new['name'] == 'PostgreSQL']
MongoDB_data = df_new[df_new['name'] == 'MongoDB']
es_data = df_new[df_new['name'] == 'Elasticsearch']
redis_data = df_new[df_new['name'] == 'Redis']
```

```python
def line_2019db() -> Line:
    c = (
        Line()
        .add_xaxis(oracle_data.iloc[-12:-1]['date'].values.tolist())
        .add_yaxis("Oracle", oracle_data.iloc[-12:-1]['value'].values.tolist())
        .add_yaxis("MySQL", mysql_data.iloc[-12:-1]['value'].values.tolist())
        .add_yaxis("Microsoft SQL Server", MSSQL_data.iloc[-12:-1]['value'].values.tolist())
        .add_yaxis("PostgreSQL", Pg_data.iloc[-12:-1]['value'].values.tolist())
        .add_yaxis("MongoDB", MongoDB_data.iloc[-12:-1]['value'].values.tolist())
        .add_yaxis("Elasticsearch", es_data.iloc[-12:-1]['value'].values.tolist())
        .add_yaxis("Redis", redis_data.iloc[-12:-1]['value'].values.tolist())
        .set_global_opts(title_opts=opts.TitleOpts(title="2019流行趋势"))
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    )
    return c
```

```python
line_2019db().render_notebook()
```

```python
database = ['Oracle', 'MySQL', 'MS Server', 'PostgreSQL', 'MongoDB', 'ES', 'Redis']
database_rate = [(1336.068-1268.836)/1268.836, (1266.283-1154.267)/1154.267, (1081.907-1040.261)/1040.261, 
                 (491.071-466.114)/466.114, (413.183-387.185)/387.185, (148.402-143.441)/143.441 , (145.237-149.015)/149.015]
def funnel_db() -> Funnel:
    c = (
        Funnel()
        .add("商品", [list(z) for z in zip(database, database_rate)])
        .set_global_opts(title_opts=opts.TitleOpts(title="2019增长率"))
    )
    return c
```

```python
funnel_db().render_notebook()
```
