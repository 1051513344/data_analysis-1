```python
hot_json = json.loads(hot_data)
```

```python
hot_list = hot_json['initialState']['topstory']['hotList']
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('test.db')
```

```python
c = conn.cursor()
c.execute('''CREATE TABLE COMPANY
       (ID CHAR(20) PRIMARY KEY     NOT NULL,
       NAME           TEXT    NOT NULL,
       metrics            CHAR(20)     NOT NULL,
       link        CHAR(50),
       cardId         CHAR(20),
       );''')

conn.commit()
# conn.close()
```

```python
hot_list
```

```python
hot_list[0]['target']['metricsArea']['text']
```

```python
c = conn.cursor()
cursor = c.execute("SELECT ID from COMPANY")
hot_id = [i[0] for i in cursor]
for i in hot_list:
    if i['id'] in hot_id:
        c.execute("UPDATE COMPANY set metrics = '%s' where ID='%s'" % (i['target']['metricsArea']['text'], str(i['id'])))
        conn.commit()
    else:
        c.execute("INSERT INTO COMPANY (ID, NAME, metrics, link, cardId) VALUES ('%s', '%s', '%s', '%s', '%s')" % (i['id'], 
                                                                                                                  i['target']['titleArea']['text'],
                                                                                                                  i['target']['metricsArea']['text'],
                                                                                                                  i['target']['link']['url'], 
                                                                                                                  i['cardId']))
        conn.commit()

```
