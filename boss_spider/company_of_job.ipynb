```python
import pandas as pd
from pyecharts.charts import Bar, Geo, WordCloud, Grid, Line, Scatter, Pie
from pyecharts import options as opts
from pyecharts.globals import ChartType, SymbolType
```

```python
beijing = pd.read_csv("beijing_data.csv")
shanghai = pd.read_csv("shanghai_data.csv")
shenzhen = pd.read_csv("shenzhen_data.csv")
guangzhou = pd.read_csv("guangzhou_data.csv")
hangzhou = pd.read_csv("hangzhou_data.csv")
nanjing = pd.read_csv("nanjing_data.csv")
wuhan = pd.read_csv("wuhan_data.csv")
xian = pd.read_csv("xian_data.csv")
chengdu = pd.read_csv("chengdu_data.csv")
```

```python
# 合并城市数据
all_data = pd.concat([beijing, shanghai, shenzhen, guangzhou, hangzhou, nanjing, wuhan, xian, chengdu], ignore_index=True)
all_data
```

```python
import re
```

```python
def get_num(mystr):
    rege = r'(\d+)-(\d+)K'
    res = re.match(rege, mystr)
    result = (int(res.group(1)) + int(res.group(2)))/2
    return int(result)
```

```python
all_data['avg_salary'] = all_data['salary'].apply(get_num)
all_data
```

```python
all_company_avg_salary = all_data.groupby('company_name').mean()
all_company_avg_salary.reset_index(inplace=True)

all_company_avg_salary.sort_values(by='avg_salary', ascending=False, inplace=True)

all_job_count = all_data.groupby('company_name').size()
dict_site = {'company_name': all_job_count.index, 'counts': all_job_count.values}
all_job_count_data = pd.DataFrame(dict_site)
all_job_count_data
```

```python
all_company_avg_salary['avg_salary'] = all_company_avg_salary['avg_salary'].apply(lambda x: round(x, 2))
all_company_avg_salary_counts = pd.merge(all_company_avg_salary, all_job_count_data, on='company_name')
all_company = all_company_avg_salary_counts.sort_values(by='counts', ascending=False)
```
