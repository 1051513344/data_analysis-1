```python
provinces = ['北京', '上海', '广州', '深圳', '成都', '杭州', '南京', '武汉', '西安']
```

```python
total_values = [(float(Python_values[i]) + float(Java_values[i]) + float(dataanalyse_values[i]) + float(product_values[i]))/4 for i in range(0, len(Python_values))]
total_values
```

```python
def geo_base() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="china")
        .add("", [list(z) for z in zip(provinces, Python_values)])
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(is_piecewise=True, max_=26, min_=12),
            title_opts=opts.TitleOpts(title="Python 需求情况"),
        )
    )
    return c
```

```python
geo_base().render_notebook()
```

```python
def geo_heatmap() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="china")
        .add("", [list(z) for z in zip(provinces, total_values)], type_=ChartType.HEATMAP,)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(is_piecewise=False, max_=25, min_=12),
            title_opts=opts.TitleOpts(title="需求热力图"),
        )
    )
    return c
```

```python
geo_heatmap().render_notebook()
```

```python
def wordcloud_base(job) -> WordCloud:
    c = (
        WordCloud()
        .add("", job, word_size_range=[20, 100], shape='pentagon')
        .set_global_opts(title_opts=opts.TitleOpts(title=""))
    )
    return c
```
